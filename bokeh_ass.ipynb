A Bokeh server is a Python server application that allows you to create interactive plots and application with Bokeh. It  enables you to build web-based dashboard and appliaction that can be updated and respond to user interactions in real time

```python
#1. already install Bokeh library
```

```python
#2. import the necessary modules and functions from the bokeh library
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource
from bokeh.layouts import row
from bokeh.server.server import Server
```

```python
x = [1,2,3,4,5]
y = [6,7,3,8,4]
source = ColumnDataSource(data={'x':x, 'y':y})
```

```python
p4 = figure(title="Interactive Plot", x_axis_label = 'X',y_axis_label = 'Y')
```

```python
p4.circle('x','y', source=source, size= 10, color='green', alpha=0.5)
```

```python
def update_data():
  new_x = [1,2,3,4,5]
  new_y= [6,7,2,4,5]
  source.data = {'x': new_x,'y':new_y}
  p.on_event('tap', update_data)
```

```python
layout = row(p4)
```

```python
curdoc().add_root(layout)
```

```python
show(p4)
```

To embed a Bokeh plot into a web page or dashboard using Flask or Django,

**Using Flask**


1.  step up a flask application: Create
a flask application by initializing an instance of the 'Flask' class.
2.   Define a route: Defne a route in your flask application that corresponds to the URL where you want to dispay the Bokeh plot.

3.  Create a Bokeh plot:Use the bokrh library to create your interactive plot.
4.  Save the plot to an HTML file: Save the Bokeh plot to an HTML file using the 'bokeh.plotting.save' function.
5.  Render the HTML file in the route:In the route you defined, render the HTML file containing the Bokeh plot using Flask's 'render_template' function.
6.  Run the flask application: Start the Flask development server to run your application and access the web page containing the embedded Bokeh plot.

**Using Django**

1.  Set up a Django project:Create a Django project using the 'django_admin' command_line tool or uyour preferred method.
2.  Create a Django app: Create a Django app within your project using the python manage.py startapp command.

3.  Define a view: In your Django app, define a view function that corresponds to the URL where you want to display the Bokeh plot. This function will handle the rendering of the plot.

4.  Create a Bokeh plot: Use the Bokeh library to create your interactive plot.

5.  Save the plot to an HTML file: Save the Bokeh plot to an HTML file using the bokeh.plotting.save function.

6.  Configure the view: In the view function, load the HTML file containing the Bokeh plot and pass it as a context variable to the Django template engine.

7.  Create a Django template: Create a Django template (HTML file) that will be rendered by the view function. Embed the Bokeh plot in the template using the appropriate template tags.

8.  Define URL mapping: In your Django project's urls.py file, define a URL mapping that connects the desired URL to the view function.

9.  Run the Django server: Start the Django development server to run your application and access the web page containing the embedded Bokeh plot.
